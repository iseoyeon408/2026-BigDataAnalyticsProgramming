```python
from IPython.display import display, HTML

def display_side_by_side(*dfs):
    html_str = ''
    for df in dfs:
        html_str += df.to_html()
    display(HTML(f"<div style='display:flex; gap:20px'>{html_str}</div>"))
```

```python
values = {"A": 100, "B": 200, "C": 300, "D": 400}
df.fillna(values)
```

```python
values
```

```python
display_side_by_side(df, df.fillna(values))
```

```python
display_side_by_side(df, df.fillna(values, limit = 2))
```

```python
s = pd.Series([100, 200, 400], index = list('ACD'))
s
```

```python
print(s)
```

```python
display_side_by_side(df, df.fillna(s, limit = 1))
```

```python
display_side_by_side(df, df.fillna(s, limit = 2))
```

```python
df2 = pd.DataFrame(-np.ones((4, 4)), columns=list('ABCE'))
df2
```

```python
display_side_by_side(df, df.fillna(df2))
```
