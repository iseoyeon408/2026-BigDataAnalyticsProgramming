### 내장함수 hash()

```python
hash((1, 2))
```

```python
hash((1, 2))    #값이 나오는게 중요함!!(내용은 상관없음)
```

```python
hash([1, 2])    #hash는 수정불가능한것만 쓸수있음(리스트는 수정가능하기때문에 에러가 남!)
```

```python
hash(1)
```

```python
hash({1, 2})    #얘도 마찬가지로 set은 수정가능하기 때문에 오류가 일어남
```

```python
hash([1:100, 2:200])  #딕셔너리도 안됨
```
